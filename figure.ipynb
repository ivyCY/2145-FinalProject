```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Year": list(range(2000, 2021)),
    "WO": [101, 118, 65, 53, 72, 57, 92, 89, 44, 83, 68, 110, 124, 159, 167, 193, 262, 435, 728, 1035, 873],
    "US": [129, 264, 225, 247, 263, 323, 348, 278, 290, 303, 320, 421, 602, 836, 953, 1216, 1612, 2450, 3180, 3870, 1751],
    "CN": [7, 4, 7, 7, 10, 21, 30, 23, 35, 32, 35, 103, 192, 197, 155, 371, 959, 1998, 3101, 3681, 3818],
    "JP": [188, 249, 235, 261, 298, 283, 253, 124, 187, 93, 90, 97, 117, 148, 121, 175, 216, 273, 386, 447, 64],
    "KR": [5, 6, 10, 20, 21, 22, 31, 19, 39, 23, 38, 31, 55, 62, 56, 76, 165, 297, 557, 643, 213],
    "EP": [69, 49, 33, 33, 43, 39, 26, 28, 9, 9, 11, 22, 22, 22, 28, 50, 76, 121, 204, 270, 112],
    "IN": [0, 0, 0, 1, 0, 1, 2, 1, 1, 1, 2, 2, 5, 5, 6, 20, 25, 37, 46, 96, 119]
}

df = pd.DataFrame(data)
df = df.set_index("Year")
df['Total'] = df.sum(axis=1)
df_percentage = df.loc[:, "WO":"IN"].div(df["Total"], axis=0)

plt.figure(figsize=(14, 8))
df_percentage.plot(kind="bar", stacked=True)
plt.title("Trends in countries' patent share in AI (2000-2020)")
plt.xlabel("Year")
plt.ylabel("Percentage of patents")
plt.legend(title="country", loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xticks(rotation=45)
plt.show()
```
